```python
%%writefile mpi_init.c

#include <mpi.h>
#include <stdio.h>

int main(int argc, char *argv[]) {

  int rank, size;
  int number;

  //Initilize
  MPI_Init(&argc,&argv);

  MPI_Comm_rank(MPI_COMM_WORLD, &rank);
  MPI_Comm_size(MPI_COMM_WORLD, &size);

  if(rank==0){

    number= 100;
    printf("Process %d sending number %d to process 1\n",rank,number);
    MPI_Send(&number,1,)

  }


}

```
